```python
label_encoder = LabelEncoder()
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
x = df.drop(columns=['price'])
y = df['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_hat = model.predict(x_test)
```

```python
r2_scikit = r2_score(y_test, y_hat)
r2_scikit
```

## Implementing Linear Regression

```python
def square_trick(m, c, x, y, lr):
    y_hat = sum(w * xi for w, xi in zip(m, x)) + c
    err = y - y_hat
    m = [w + lr * xi * err for w, xi in zip(m, x)]
    c += lr * err
    
    return m, c
```

```python
def fit(X, Y, lr=0.01, epochs = 1000):
    m = list(np.random.rand(len(X[0])))
    c = np.random.rand()
    for _ in range(epochs):
        for x, y in zip(X, Y):
            m, c = square_trick(m, c, x, y, lr)
    return m, c
```

```python
def predict(m, X, c):
    return np.dot(X, m) + c
```

```python
def r2_metric(y, y_hat):
    tss = np.sum( (y - np.mean(y)) **2)
    sse = np.sum( (y - y_hat) **2)

    r2 = 1 - (sse / tss)
    return r2
```

```python
m, c = fit(x_train, y_train)
y_hat2 = predict(m, x_test, c)
r2_custom = r2_metric(y_test, y_hat2)
```

```python
print(r2_custom)
```

## Models Comparison

```python
print(f"Scikit-learn R2: {r2_scikit}\nLinearRegression Implementation R2: {r2_custom}")
```
